```python
import pandas as pd
import numpy as np

filename = r"D:\Jules\Professionel\Stages\2024_SFE_Supaero\P2_BCI_Timeflux\timeflux_git\In-time-we-flux\data\P1_Fast_100%_95%.hdf5"
store = pd.HDFStore(filename, mode="r")
```

```python
df = store['filtered']
df.shape
```

```python
# slice df in 300ms windows
window_size = 300
step = 50
windows = []
for i in range(0, len(df)//2, step):
    windows.append(df.iloc[i:i+window_size])
# COnvert windows to dataframe
windows = [pd.DataFrame(window) for window in windows]
print(len(windows))
# Mean time series 
np.mean(windows, axis=(0,2)).shape
```

```python
# Plot the mean time series
import matplotlib.pyplot as plt
plt.plot(np.mean(windows, axis=(0,2)))
plt.show()
```

```python
# Transform windows to numpy array
windows = np.array(windows)
windows.mean(axis=(0,2)).shape

plt.plot(windows.std(axis=(0,2)))
plt.show()
```
